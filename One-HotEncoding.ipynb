```python
import pandas as pd
```

```python
companies_df = pd.read_csv('companies_information.csv')
```

```python
industries_df = pd.read_csv('industries_information.csv')
```

```python
balance_df = pd.read_csv('balance_sheet.csv')
```

```python
balance_df = balance_df.drop(columns=['Report Date',
                                      'Currency',
                                     'Publish Date',
                                     'Restated Date',
                                     'Shares (Basic)',
                                     'Shares (Diluted)'])
```

```python
cashflow_df = pd.read_csv('cashflow_statement.csv')
```

```python
print(cashflow_df.columns.tolist())
```

```python
cashflow_df = cashflow_df.drop(columns=['Report Date',
                                        'Currency',
                                        'Publish Date',
                                        'Restated Date',
                                        'Shares (Basic)',
                                        'Shares (Diluted)',
                                        'Depreciation & Amortization'])
```

```python
income_df = pd.read_csv('income_statement.csv')
```

```python
eps_df = pd.read_csv('eps_df.csv')
```

```python
merge_df = companies_df.merge(industries_df, on='IndustryId', how='left')\
                       .merge(income_df, on=['Ticker', 'SimFinId'], how='left')\
                       .merge(balance_df, on=['Ticker', 'SimFinId', 'Fiscal Year', 'Fiscal Period'], how='left')\
                       .merge(cashflow_df, on=['Ticker', 'SimFinId', 'Fiscal Year', 'Fiscal Period'], how='left')\
                       .merge(eps_df, on=['Ticker', 'Fiscal Year', 'Fiscal Period'], how='left')
```

```python
length = len(merge_df.columns)
```

```python
amazon_df = merge_df[merge_df['Ticker']=='AMZN'].reset_index(drop=True)
```

```python
amazon_df
```

```python
fiscal_encoded = pd.get_dummies(amazon_df['Fiscal Period'], dtype=int)
```

```python
fiscal_encoded
```

```python
amazon_df = amazon_df.drop('Fiscal Period', axis = 1)
```

```python
amazon_df = pd.concat([amazon_df, fiscal_encoded], axis = 1)
```
