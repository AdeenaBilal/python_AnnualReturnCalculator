```python
apple_df.fillna({'Net Cash from Acquisitions & Divestitures': apple_df['Net Cash from Acquisitions & Divestitures'].mean()}, inplace = True)
```

```python
apple_df.info()
```

```python
apple_df['Report Date_x'] = pd.to_datetime(apple_df['Report Date_x'], format ='%m/%d/%Y')
```

```python
apple_df.info()
```

```python
apple_df = apple_df.sort_values(by = 'Report Date_x').reset_index(drop=True)
```

```python
apple_df
```

```python
apple_df['Revenue' + ' change from last quarter'] = apple_df['Revenue'].pct_change()
```

```python
#apple_df
```

```python
featureToGetPercentChange = ['Shares (Basic)_x',
                             'Shares (Diluted)_x',
                             'Operating Income (Loss)',
                             'Net Income',
                             'Total Liabilities & Equity']
                             
                             
                             
```

```python
for feature in featureToGetPercentChange:
    apple_df[feature + ' change from last quarter'] = apple_df[feature].pct_change()
```

```python
#apple_df
```

```python
apple_df['Next quarter EPS'] = apple_df['Earnings Per Share, Basic'].shift(-1)
```

```python
#apple_df
```

```python
apple_df = apple_df.dropna(subset=['Next quarter EPS']).reset_index(drop=True)
```

```python
#apple_df
```

```python
apple_df = apple_df.dropna().reset_index(drop = True)
```

```python
#apple_df
```

```python
def regression_label(nextQuarter, currentQuarter):
    if abs(currentQuarter) == 0:
        return 0
    return (nextQuarter - currentQuarter)/abs(currentQuarter)

apple_df['Change in EPS (Regression_label)']= apple_df.apply(lambda row: regression_label(row['Next quarter EPS'], row['Earnings Per Share, Basic']), axis=1)
```

```python
apple_df
```
